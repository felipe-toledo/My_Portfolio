### **Q2.What is the percentage of missing values in the dataset?**
The code to know what values are **null**, you can use the "isnull" function. If the value exists, it shows you an zero, if not, it shows you as a binary number of 1. So, the number printed in the table represent the sum of the missing values for each variable. 

```
#Which variables are missing?
london.isnull()
```

So, if the function "isnull" shows TRUE, it means that the variable does NOT exist. If False, the variable exists. I know it works quite opposite interpretation, but the answer is responding the question made by which variable is null (not actually zero because "zero" still a number, but if the guests filled those information in the system).

```
london.isnull().sum()
```

Now, translating it into percentage, we divide the variable by the total number of the population, which in this case is shape [0].

```
((london.isnull().sum()/london.shape[0])*100)
```

So, after the results of the percentages for each variables, we should organize the table in an ascending form. We can see that the neighbouhood group is a total missing value. Also, the reviews per month and the last review are 19.89% of missing values. 

```
# sort in descending order the variables by their missing values
((london.isnull().sum()/london.shape[0])*100).sort_values(ascending=False)
```

Based on the results, we can infer the following:

* The neighbourhood_group column does not have any values. Therefore, it is equal to 100%. 
* The columns reviews_per_month and  last_review have the same percentage. It means that the number of reviews could be the last review for each ID account.

Therefore, we'll create a new dataframe, *london_clean* that contains the following:
* The price rows will be between Â£0 and Â£600;
* The minimum nights are no more than 30;

After that, with this new dataframe created, we can the reviews_per_month and last_review columns as we said earlier in the projectas they are the same percentage of other review's parameters.

Last but not least, the next histograms might look better.

```
london_clean = london.copy()
# Taking out values above Â£600
london_clean.drop(london_clean[london_clean.price > 600].index, axis=0, inplace=True)

# Taking out values below 30 from the column 'minimum_nights'
london_clean.drop(london_clean[london_clean.minimum_nights > 30].index, axis=0, inplace=True)

# Delete values "Zero" from the column 'availability_365'
london_clean.drop(london_clean[london_clean.availability_365 == 0].index, axis=0, inplace=True)

# Delete column 'neighbourhood_group' 
london_clean.drop('neighbourhood_group', axis=1, inplace=True)

#plot the histogram for the cleaned variables
london_clean.hist(bins=30, figsize=(20,15));
```

Now, we can see that the price appears more realistic, resembling a bell-shape curve. The number of the minimum nights do concentrate more in the range between 1 an 3 days, plunging 4 days onwards. There are two small "bumps" that could represent the rentals for 15 days and 30 days, respectively. A plausible explanation is that those ones are related with corporation's bookings in order to save more money allocating groups of contractors in entire houses rather than in a hotel industry. Also, the availability does show that the most variable is throughout of the year. Before the outliears cleaning, it was not possible to spot it.
